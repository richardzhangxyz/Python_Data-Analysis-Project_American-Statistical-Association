### 1. Find proportion of people overall in the USA who are food insecure (overall)

```python
import numpy as np
import pandas as pd
import pylab

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
#import dataset again
food_whole = pd.read_csv('http://www.stat.uchicago.edu/~nicolae/118/food_access_research_atlas_subset.csv')
print(food_whole.shape)
food_whole.head(3)
```

```python
#find total population across all counties
whole_pop = food_whole["Pop2010"].sum()
print("Total population is", whole_pop, "individuals")

#find total population count beyond 1 mile from supermarket
pop_beyond_one = food_whole["lapop1"].sum()
pop_beyond_one = round(pop_beyond_one)
print("Total population beyond 1 mile from supermarket is", pop_beyond_one, "individuals")

#find overall proportion of people in US who are insecure
prop_insecure = pop_beyond_one / whole_pop
print("Proportion of people (USA) who are food insecure is", prop_insecure)
prop_insecure_percent = (pop_beyond_one/whole_pop)*100
print("Percent of people (USA) who are food insecure is", prop_insecure_percent, "% of individuals")
```

```python
#place food secure and food secure populations together
food_sec = "Food Secure", "Food Insecure"
food_sec_quant = [60.2, 39.8]

#create pie chart to display data
figure, food_sec_graph = plots.subplots()
explode = (0, 0.15) 
plots.title("Food Security for People Overall in the USA (By %)")
food_sec_graph.pie(food_sec_quant, explode = explode, labels = food_sec, autopct = "%1.1f%%", startangle = 250)
plots.show()
```

### 3. Create 1st Metric for State Rank (50/50 Weight between population beyond one mile and low income population beyond one mile)

### 3 (A) Find population beyond one mile

```python
#import dataset again
food_one = pd.read_csv('http://www.stat.uchicago.edu/~nicolae/118/food_access_research_atlas_subset.csv')
print(food_one.shape)
food_one.head(3)
```

```python
#drop columns with NaN values (Low income population count beyond 1 mile from supermarket)
#food_one.dropna(subset = ["lapop1"], inplace = True)

#sum population and low income access population (>1 mile) by state
one_state_sum = food_one.groupby("State").sum()
one_state_sum

#find percentage of low income access population (>1 mile) by state
state_one_mile = one_state_sum["lapop1"] / one_state_sum["Pop2010"]

#append new column to dataframe
one_state_sum["state_one_mile"] = state_one_mile

#list states with highest food insecure population (>1 mile) in descending order
one_mile = one_state_sum.sort_values(by = ["state_one_mile"], ascending = False)[["state_one_mile"]]
one_mile.head(5)
```

### 3 (A) Find low income population beyond one mile

```python
#import dataset again
food_one_low = pd.read_csv('http://www.stat.uchicago.edu/~nicolae/118/food_access_research_atlas_subset.csv')
print(food_one_low.shape)
food_one_low.head(3)
```

```python
#sum population and low income access population (>1 mile) by state
one_low_state_sum = food_one_low.groupby("State").sum()
one_low_state_sum

#find percentage of low income access population (>1 mile) by state
state_low_income_one_mile = one_low_state_sum["lalowi1"] / one_low_state_sum["Pop2010"]

#append new column to dataframe
one_low_state_sum["state_low_income_one_mile"] = state_low_income_one_mile

#list states with highest low income access population (>1 mile) in descending order
one_mile_low = one_low_state_sum.sort_values(by = ["state_low_income_one_mile"], 
                                             ascending = False)[["state_low_income_one_mile"]]
one_mile_low.head(5)
```

### 3 (A) Create DataFrame for population beyond one mile and low income population beyond one mile

```python
#merge two dataframes above
one_and_low = pd.merge(one_mile, one_mile_low, left_on = "State", right_on = "State")

#weigh 50/50 for both categories
one_weight = (one_and_low["state_one_mile"] * 50)
one_low_weight = (one_and_low["state_low_income_one_mile"] * 50)

#create a comprehensive weight between two
weighted_one_metric = one_weight + one_low_weight

#append as new column in dataframe
one_and_low["Weighted Metric"] = weighted_one_metric

#rank states in descending order
one_and_low = one_and_low.sort_values(by = ["Weighted Metric"], ascending = False)[["Weighted Metric"]]
one_and_low
```

```python
#list 5 worst states
one_and_low
one_and_low.head(5).style.set_caption("Top 5 States with Worst Food Security Based on First Metric Computation")
```

```python
#list 5 best states
one_and_low.tail(5).style.set_caption("Top 5 States with Best Food Security Based on First Metric Computation")
```

### 3. Create 2nd Metric for State Rank (50/50 Weight between low income population beyond one mile and low income population beyond ten miles)

### 3 (A) Find low income population beyond ten miles

```python
#import dataset again
food_ten_low = pd.read_csv('http://www.stat.uchicago.edu/~nicolae/118/food_access_research_atlas_subset.csv')
print(food_ten_low.shape)
food_ten_low.head(3)
```

```python
#drop columns with NaN values (Low income population count beyond 10 miles from supermarket)
#food_ten_low.dropna(subset = ["lalowi10"], inplace = True)

#sum population and low income access population (>10 miles) by state
ten_low_state_sum = food_ten_low.groupby("State").sum()

#find percentage of low income access population (>1 mile) by state
ten_low_state_sum_percent = ten_low_state_sum["lalowi10"] / ten_low_state_sum["Pop2010"]

#append new list to dataframe
ten_low_state_sum["ten_low_state_sum_percent"] = ten_low_state_sum_percent

#list states with highest low income access population (>1 mile) in descending order
ten_mile_low = ten_low_state_sum.sort_values(by = ["ten_low_state_sum_percent"], 
                                             ascending = False)[["ten_low_state_sum_percent"]]
ten_mile_low.head(5)
```

```python
#merge two dataframes above
one_ten = pd.merge(ten_mile_low, one_mile_low, left_on = "State", right_on = "State")

#weigh 50/50 for both categories
ten_weight = (one_ten["ten_low_state_sum_percent"] *50)
one_weight = (one_ten["state_low_income_one_mile"] *50)

#create a comprehensive weight between two
weighted_metric = ten_weight + one_weight

#append as new column in dataframe
one_ten["Weighted Metric"] = weighted_metric

#rank states in descending order
one_ten = one_ten.sort_values(by = ["Weighted Metric"], ascending = False)[["Weighted Metric"]]
```

### 4. Find correlation between tract poverty rate with food access

```python
#import dataset
food_poverty = pd.read_csv("Food Access Research Atlas.csv")
print(food_poverty.shape)
food_poverty.head(5)
```

```python
#create new set of percentage values for poverty rate
poverty_percent = food_poverty["PovertyRate"] / 100

#insert poverty_percent as a new column in the original dataframe
food_poverty["poverty_percent"] = poverty_percent
food_poverty

#find quantity of individuals who are in poverty in each county
poverty_quant = food_poverty["Pop2010"] * food_poverty["poverty_percent"]
food_poverty["poverty_quant"] = poverty_quant

#groupby states
state_poverty_sum = food_poverty.groupby("State").sum()
state_poverty_sum

#find total quantity of individuals in poverty
total_poverty = (state_poverty_sum["poverty_quant"] / state_poverty_sum["Pop2010"])
state_poverty_sum["total_poverty_percent"] = total_poverty

#define and show first 5 rows
poverty_percentage = state_poverty_sum[["total_poverty_percent"]]
poverty_percentage.head(5)
```

```python
#create dataframe merged between state poverty rate and state food security rate
correlation_pov_access = pd.merge(poverty_percentage, one_mile_low, left_on = "State", right_on = "State")
correlation_pov_access.head(5)
```

```python
#plot scatterplot to display correlation
correlation_pov_access.plot.scatter(x = "state_low_income_one_mile", y = "total_poverty_percent")
plots.xlabel("Percentage of Individuals who are Food Insecure")
plots.ylabel("Poverty Rate")
plots.title("Correlation: Food Insecurity and Poverty Rate (By State)")
```
